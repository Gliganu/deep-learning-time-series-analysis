```python
output = np.zeros((similarity.shape[0], 3))

identical = 0
above_99 = 0

for i in range(100):
    if similarity[i][:].max() == 1:
        identical += 1
    else:
        if similarity[i][:].max() >= 0.99:
            above_99 += 1
    output[i][0] = unknown_pids[i]
    output[i][1] = train_pids[similarity[i][:].argmax()]
    output[i][2] = similarity[i][:].max()

print('Identical: %d' % identical)
print('Similarity above 0.99: %d' % above_99)

output_df = pd.DataFrame(output, columns=['unseen_pid', 'train_pid', 'similarity'])
output_df[['unseen_pid', 'train_pid']] = output_df[['unseen_pid', 'train_pid']].astype(int)
```
