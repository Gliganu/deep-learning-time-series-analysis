```python
joined['prevPid'] = joined['pid'].shift(1)
joined['predvDay'] = joined['day'].shift(1)
joined['prevDailyOccurence'] = joined['dailyOccurence'].shift(1)
joined['nextLineID'] = joined['lineID'].shift(-1)
```

```python
joined
```

```python
def apply(x):
    if x['prevPid'] == x['pid'] and x['day'] == x['predvDay']:
        joined.loc[x.name + 1, 'prevDailyOccurence'] = x['prevDailyOccurence'] + 1
        joined.loc[x.name, 'dailyOccurence'] = x['prevDailyOccurence'] + 1
    else:
        joined.loc[x.name + 1, 'prevDailyOccurence'] = 0
        joined.loc[x.name, 'dailyOccurence'] = 0
joined.apply(lambda x: apply(x), axis = 1)
```

```python
prev_day = 0
prev_pid = 0
current_occurence = 0

for t in joined.itertuples():
    if t.pid != prev_pid or t.day != prev_day:
        print('pid: %d day: %d' %(t.pid, t.day))
        current_occurence = 0
    joined.loc[t.Index, 'dailyOccurence'] = current_occurence
    current_occurence += 1
    prev_pid = t.pid
    prev_day = t.day
```
